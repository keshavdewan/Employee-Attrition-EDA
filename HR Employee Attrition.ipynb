## Visualization & Determining Variables

#### Education

```python
result_education = []

for value in df['Education']:
    if value == 1:
        result_education.append('Below College')
    elif value == 2:
        result_education.append('College')
    elif value == 3:
        result_education.append('Bachelor')
    elif value == 4:
        result_education.append('Master')
    else:
        result_education.append('Doctor')
```

```python
result_education
```

```python
df['Education'] = result_education
```

```python
df['Education'].value_counts(dropna = False)
```

```python
# Bar chart Education

education_df = df['Education'].value_counts().sort_values()

fig,ax = plt.subplots(figsize = (10,4), dpi = 90)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(True)
ax.tick_params(bottom=False)
ax.get_yaxis().set_visible(False)

sns.countplot(data=df, x='Education');

# add values on top of each bar
ax.bar_label(ax.containers[0])

ax.set_xlabel('Education', fontsize=13, color = '#ff4800');
fig.suptitle('Employee Education', fontsize=15, color = '#ff4800')
```

Most of the employees have a Bachelor's followed by a Masters degree

#### Attrition Rate

```python
#Checking Attrition rate (%) 

# Pie chart

pie_df = df['Attrition'].value_counts
fig, (ax) = plt.subplots(1, figsize = (10,5))

pie_bar_colors = ['#FB8E7E', '#8EC9BB']
explode = [0,0.1]

ax.pie(df['Attrition'].value_counts().values, labels = df['Attrition'].value_counts().index,
       explode = explode, colors = pie_bar_colors, autopct = '%1.1f%%')
ax.axis('equal')

fig.suptitle('Attrition', fontsize=15, color = '#ff4800')

```

The organisation has a very less attrition rate of 16%

```python
#Checking Business Travel (%) 

# Pie chart

pie_df = df['Business_Travel'].value_counts
fig, (ax) = plt.subplots(1, figsize = (10,5))

pie_bar_colors = ['#FB8E7E', '#8EC9BB','#00FF00']
explode = [0,0.1]

ax.pie(df['Business_Travel'].value_counts().values, labels = df['Business_Travel'].value_counts().index,
      colors = pie_bar_colors, autopct = '%1.1f%%')
ax.axis('equal')

fig.suptitle('Business Travel', fontsize=15, color = '#ff4800')
```
